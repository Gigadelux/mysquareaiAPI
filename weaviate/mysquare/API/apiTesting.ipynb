```python
import requests

url = "http://localhost:8000/get_people/"
params = {
    "prompt": "Hi! I want to find people that loves improve themselves and fitness ones, that ate low calory foods",
    "apikey": "dksjnrgsdgsd"
}

response = requests.get(url, params=params)

if response.status_code == 200:
    print("Success!")
    print(response.json())
else:
    print("Failed with status code:", response.status_code)
    print(response.text)
```
